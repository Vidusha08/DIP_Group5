```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
from PIL import Image, ImageTk
import cv2
import copy
import numpy as np
import matplotlib.pyplot as plt
from cv2 import dnn_superres
import tensorflow as tf

#create tkinter windwo
root = tk.Tk()
root.title("Application")

frame = ttk.Frame(root)
frame.pack(fill='both', expand=True)

# Create a canvas for the scrollbar
canvas = tk.Canvas(frame)
canvas.pack(side='left', fill='both', expand=True)

# Create a scrollbar
scrollbar = ttk.Scrollbar(frame, orient='vertical', command=canvas.yview)
scrollbar.pack(side='right', fill='y')
canvas.configure(yscrollcommand=scrollbar.set)

# Create a frame to hold the content (e.g., image, text, etc.)
content_frame = ttk.Frame(canvas)
canvas.create_window((-1, -1), window=content_frame, anchor='w')

content_frame.update_idletasks()
canvas.config(scrollregion=canvas.bbox('all'))

# Function to scroll the canvas
def on_mousewheel(event):
    canvas.yview_scroll(int(-1 * (event.delta / 120)), 'units')

canvas.bind_all("<MouseWheel>", on_mousewheel)


img = None
original_img = None

brightess = 1.0
contrast = 1.0

#function to load an image
def load_image():
    global img, original_img
    file_path = filedialog.askopenfilename()
    if file_path:
        img = cv2.imread(file_path)
        original_img = copy.deepcopy(img)
        display_image(img, file_path)
        
#funciton to display the image
def display_image(img, file_path):
    if img is not None:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        image = Image.fromarray(img)
        photo = ImageTk.PhotoImage(image=image)
        image_label.config(image=photo)
        image_label.image = photo
        file_path_label.config(text="Image Path: " + file_path)
    else:
        properties_label.config(text="Failed to display the image.")
            
#function to acces properties
def access_properties():
    global img
    if img is not None:
        height, width, channels = img.shape
        properties_label.config(text = "Image Properties: Height - " + str(height) + ", Width - " + str(width) + ", Channels - " + str(channels))
    else:
        properties_label.config(text = "Image not loaded.")

#function to flip the image horizontaly
def flip_hori():
    global img
    if img is not None:
        flipped_image = cv2.flip(img, 1)
        display_image(flipped_image, "Horionztaly Flipped Image")
        properties_label.config(text = "Horionztaly Flipped Image")
    else:
        properties_label.config(text = "No Image")
        
#function to flip the image vertically
def flip_vert():
    global img
    if img is not None:
        flipped_image = cv2.flip(img, 0)
        display_image(flipped_image, "Vertically Flipped Image")
        properties_label.config(text = "Vertically Flipped Image")
    else:
        properties_label.config(text = "No Image")
        
#fucntion to rotate the image
def rotate_image(degree):
    global img
    if img is not None:
        try:
            degrees = int(degree)
            height, width = img.shape[:2]
            center = (width // 2, height // 2) 
            rotationMatrix = cv2.getRotationMatrix2D(center, degrees, 1)
            img = cv2.warpAffine(img, rotationMatrix, (width, height))
            display_image(img, "Rotated Image")
            properties_label.config(text = "Rotated Image")
        except ValueError:
            properties_label.config(text = "Invalid Rotation Degree.")
        
#fucntion to resize the image
def resize_image(new_width, new_height):
    global img
    if img is not None:
        try:
            new_width = int(new_width)
            new_height = int(new_height)
            img = cv2.resize(img, (new_width, new_height))
            display_image(img, "Resized Iamge")
            properties_label.config(text = "Resized Image")
        except ValueError:
            properties_label.config(text = "Invalid Width or Height.")
            
#function to crop the image
def crop_image(x1,y1,x2,y2):
    global img 
    if img is not None:
        try:
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)
            cropped_img = img[y1:y2, x1:x2]
            display_image(cropped_img, "Cropped Image")
            properties_label.config(text = "Cropped Image")
        except ValueError:
            properties_label.config(text = "Invalid coordinated for cropping.")
            
#function to edge detection
def edge_detection():
    global img
    if img is not None:
        detected_img = cv2.Canny(img, 100, 200)
        cv2.imshow("Edge Ditected Image", detected_img)
        cv2.waitKey(0)
     

#function to Smoothing the image 
def smooth_image():
    global img
    if img is not None:
        kernel_size = 5
        sigma = 0
        kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / (kernel_size**2)
        smoothed_image = cv2.filter2D(img, -1, kernel)
        cv2.imshow("Smoothed Image", smoothed_image)
        cv2.waitKey(0)
    
        
#function to sharpen the image
def sharpen_image():
    global img
    if img is not None:
        kernel = np.array([[-1, -1, -1],
                           [-1, 9, -1],
                           [-1, -1, -1]])
        sharpened_image = cv2.filter2D(img, -1, kernel)
        cv2.imshow("Sharpened Image", sharpened_image)
        cv2.waitKey(0)
    
    
#function to emboss the loaded image
def emboss_image():
    global img
    if img is not None:
        kernel = np.array([[0, 1, 0],
                         [0, 0, 0],
                         [0, -1, 0]])
        embossed_image = cv2.filter2D(img, -1, kernel)
        cv2.imshow("Embossed Image", embossed_image)
        cv2.waitKey(0)
    

#function to adjust the brightness of the image
def brightness_adjust():
    global brightness
    if img is not None:
        try:
            brightness = float(brightness_entry.get())
            adjusted_image = cv2.convertScaleAbs(img, alpha = brightness)
            cv2.imshow('Adjusted Image', adjusted_image)
        except ValueError:
            pass  
    
#function to adjust the contrast of the image
def contrast_adjust():
    global contrast
    if img is not None:
        try:
            contrast = float(contrast_entry.get())
            adjusted_image = cv2.convertScaleAbs(img, beta = contrast)
            cv2.imshow('Adjusted Image', adjusted_image)
        except ValueError:
            pass  
        
#function to enhance the image
def enhance_img():
    global img
    sr = dnn_superres.DnnSuperResImpl_create()
    path = "EDSR_x3.pb"
    sr.readModel(path)
    sr.setModel("edsr", 3)
    result = sr.upsample(img)
    cv2.imshow("Enhanced Image", result)
    
def load_and_preprocess_image(image_path):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img

# Function to apply style transfer.
def style_transfer(content_path, style_path, num_iterations=1000, content_weight=1e3, style_weight=1e-2):
    content_image = load_and_preprocess_image(content_path)
    style_image = load_and_preprocess_image(style_path)

    # Define the VGG model with style layers
    style_layers = ['block1_conv1', 'block2_conv1', 'block3_conv1', 'block4_conv1', 'block5_conv1']
    vgg = tf.keras.applications.VGG19(include_top=False, weights='imagenet')
    style_model = tf.keras.models.Model(inputs=vgg.input, outputs=[vgg.get_layer(layer).output for layer in style_layers])

    # Calculate style targets
    style_targets = style_model(style_image)

    generated_image = tf.Variable(content_image)

    opt = tf.optimizers.Adam(learning_rate=0.02, beta_1=0.99, epsilon=1e-1)

    return generated_image
    
generated_image_tk = None

def open_and_apply_style_transfer():
    content_path = filedialog.askopenfilename(title="Select Content Image")
    if not content_path:
        return
    style_path = filedialog.askopenfilename(title="Select Style Image")
    if not style_path:
        return
    
    # Call style transfer function
    generated_image = style_transfer(content_path, style_path)
    
    # Create a label to display the generated image
    style_label = ttk.Label(content_frame)
    style_label.pack()
    
    # Convert the generated image to a format that Tkinter can display
    generated_image = (generated_image[0].numpy() * 255).astype(np.uint8)
    generated_image = cv2.cvtColor(generated_image, cv2.COLOR_BGR2RGB)
    generated_image = Image.fromarray(generated_image)
    generated_image_tk = ImageTk.PhotoImage(generated_image)
    
    # Update the label with the generated image
    style_label.config(image=generated_image_tk)
    style_label.image = generated_image_tk

                          

#function to undo all changes
def undo():
    global img, original_img
    if original_img is not None:
        img = copy.deepcopy(original_img)
        display_image(img, "Original Image")
        properties_label.config(text = "Undo")

        
#create application UI

image_label = tk.Label(content_frame)
image_label.pack()

load_image_btn = ttk.Button(content_frame, text = "Load Image", command = load_image)
load_image_btn.pack()

file_path_label = tk.Label(content_frame, text = "Image Path: ")
file_path_label.pack()

properties_btn = ttk.Button(content_frame, text="Access Image Properties", command = access_properties)
properties_btn.pack()

properties_label = tk.Label(content_frame, text = "")
properties_label.pack()

hori_flip_btn = ttk.Button(content_frame, text = "Flip Horizontally", command = flip_hori)
hori_flip_btn.pack()

vert_flip_btn = ttk.Button(content_frame, text = "Flip Vertically", command = flip_vert)
vert_flip_btn.pack()

rotation_label = ttk.Label(content_frame, text = "Enter Rotation Degree: ")
rotation_label.pack()

rotation_entry = ttk.Entry(content_frame)
rotation_entry.pack()

rotate_btn = ttk.Button(content_frame, text = "Rotate Image", command = lambda: rotate_image(rotation_entry.get()))
rotate_btn.pack()

new_widht_label = ttk.Label(content_frame, text = "Enter New Widht: ")
new_widht_label.pack()

new_widht_entry = ttk.Entry(content_frame)
new_widht_entry.pack()

new_height_label = ttk.Label(content_frame, text = "Enter New Height: ")
new_height_label.pack()

new_height_entry = ttk.Entry(content_frame)
new_height_entry.pack()

resize_btn = ttk.Button(content_frame, text = "Resize Image", command = lambda: resize_image(new_widht_entry.get(), new_height_entry.get()))
resize_btn.pack()

x1_label = ttk.Label(content_frame, text="x1:")
x1_label.pack()

x1_entry = ttk.Entry(content_frame)
x1_entry.pack()

y1_label = ttk.Label(content_frame, text="y1:")
y1_label.pack()

y1_entry = ttk.Entry(content_frame)
y1_entry.pack()

x2_label = ttk.Label(content_frame, text="x2:")
x2_label.pack()

x2_entry = ttk.Entry(content_frame)
x2_entry.pack()

y2_label = ttk.Label(content_frame, text="y2:")
y2_label.pack()

y2_entry = ttk.Entry(content_frame)
y2_entry.pack()

crop_button = ttk.Button(content_frame, text="Crop Image", command = lambda: crop_image(x1_entry.get(), y1_entry.get(), x2_entry.get(), y2_entry.get()))
crop_button.pack()

emboss_btn = ttk.Button(content_frame, text = "Emboss Image", command = emboss_image)
emboss_btn.pack()

smooth_btn = ttk.Button(content_frame, text = "Smoothen Image", command = smooth_image)
smooth_btn.pack()

sharpen_btn = ttk.Button(content_frame, text = "Sharpen Image", command = sharpen_image)
sharpen_btn.pack()

edge_detection_btn = ttk.Button(content_frame, text = "Detect Edges of the Image", command = edge_detection)
edge_detection_btn.pack()

brightness_label = ttk.Label(content_frame, text=" Enter Brightness Level:")
brightness_label.pack()

brightness_entry = ttk.Entry(content_frame)
brightness_entry.pack()

brightness_btn = ttk.Button(content_frame, text = "Adjust Image Brightness", command = brightness_adjust)
brightness_btn.pack()

contrast_label = ttk.Label(content_frame, text=" Enter Contrast Level:")
contrast_label.pack()

contrast_entry = ttk.Entry(content_frame)
contrast_entry.pack()

contrast_btn = ttk.Button(content_frame, text = "Adjust Image Contrast", command = contrast_adjust)
contrast_btn.pack()

enhance_btn = ttk.Button(content_frame, text = "Enhance Image", command = enhance_img)
enhance_btn.pack()
                                                          
style_transfer_button = ttk.Button(content_frame, text="Apply Style Transfer", command=open_and_apply_style_transfer)
style_transfer_button.pack()

undo_btn = ttk.Button(content_frame, text = "Undo Changes", command = undo)
undo_btn.pack()

#start the tkinter window
root.mainloop()
```
